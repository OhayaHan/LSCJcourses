### 2. 数据分析题
本次课程的在线资源文件夹中包含了一个名为countries_data.csv的文件，其中中收集了30个国家的两项指标，这些国家被分成了两类（label 1 和 label 0），请你建立LR模型对这两类国家进行划分。

具体要求：  
1）请使用Python或者R完成。  
2）提交完整的源代码（包括你引用头文件所需要的代码），以及你程序的输出结果。如果你的程序无法正确执行或者结果错误，你可能会被黄牌扣罚。  
3）评估你的模型（用本次课程中学到的指标对留存数据集进行评估）。

```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries_data.csv',encoding='gbk')
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,1:3]
Y = df.iloc[:,3]

#X.as_matrix()
#FutureWarning: Method .as_matrix will be removed in a future version. Use .values instead.
X = X.values
Y = Y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lm = linear_model.LogisticRegression(C=1e5,  solver='lbfgs', multi_class='multinomial')
```

```python
lm.fit(X_train, y_train)
```

```python
prediction = lm.predict(X_test)
```

```python
print("accuracy score: ")
print(accuracy_score(y_test, prediction))
print(classification_report(y_test, prediction))
```

```python
df = pd.read_excel('TreeAge_Height.xlsx')
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,0].values
Y = df.iloc[:,1].values
```

```python
#X的平均值
xm = X.mean()
#Y的平均值
ym = Y.mean()
xm, ym
```

```python
def f(X,Y):
    beta1 = (sum(X * Y) - 24 * xm * ym)/(sum(X*X) - 24 * xm * xm)
    beta0 = ym - beta1 * xm
    return beta0, beta1
```

```python
f(X,Y)
```
