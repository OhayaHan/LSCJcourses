# Q1 统计朝阳医院的每个月的销售数量和应收金额总和，并绘制成条形图和折线图


```python
import pandas as pd
import numpy as np
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
```

```python
df = pd.read_excel('朝阳医院2018年销售数据.xlsx')
df.head()
```

```python
df['购药日期'] = df['购药时间'].str.split(' ').str.get(0)
df = df[df['购药日期']!='2018-02-29']
df['购药日期'] = pd.to_datetime(df['购药时间'].str.split(' ').str.get(0), format='%Y-%m-%d')
df.set_index('购药日期', inplace=True)
df.head()
```

```python
df.to_period('1m')
```

```python
df['month'] = df.to_period('1m').index
```

```python
dfg = df.groupby('month')['销售数量', '应收金额'].sum()
```

```python
dfg.head()
```

```python
def bar_base() -> Bar:
    c = (
        Bar()
        .add_xaxis(['2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
             '2018-07'])
        .add_yaxis("销售数量", dfg['销售数量'].to_list())
        .set_global_opts(title_opts=opts.TitleOpts(title="每个月的销售数量", subtitle="sub"))
    )
    return c
bar_base().render_notebook()
```

```python
def bar_base() -> Bar:
    c = (
        Bar()
        .add_xaxis(['2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
             '2018-07'])
        .add_yaxis("应收金额", dfg['应收金额'].map(int).to_list())
        .set_global_opts(title_opts=opts.TitleOpts(title="每个月的应收金额", subtitle="sub"))
    )
    return c
bar_base().render_notebook()
```
