```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 支持中文显示
plt.rcParams['font.sans-serif'] = ['STFangsong']
plt.rcParams['axes.unicode_minus'] = False
import os
```

## Q1 利用所学的子图相关知识，将朝阳医院不同商品的销售数量和销售总额分别进行绘制，显示在两张子图上面

```python
df = pd.read_excel('朝阳医院2018年销售数据.xlsx')
df.head()
```

```python
s1 = df.groupby('商品名称')['销售数量'].sum()
```

```python
s2 = df.groupby('商品名称')['实收金额'].sum()
```

```python
# 设置绘图区域
plt.figure(figsize=(20, 10))
plt.subplot(121) # 位置
plt.plot(s1.index,s1.values,c= 'c')
plt.xlabel("商品名称",fontsize=18)
plt.ylabel('销售数量',fontsize=18)
plt.xticks(rotation=45)
plt.subplot(122)
plt.plot(s2.index,s2.values,c= 'c')
plt.xlabel("商品名称",fontsize=18)
plt.ylabel('实收金额',fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

## Q2根据所学的子图相关知识，统计在sec_cars数据集里面，统计不同品牌汽车的平均价格，在taobao_data数据集中，统计不同位置的成交总价和总量，分别绘制在三张分布图上面



```python
df = pd.read_csv('sec_cars.csv')
df.head()
```

```python
df1 = pd.read_csv('taobao_data.csv')
df1.head()
```

```python
s1 = df.groupby('Brand')['Sec_price'].mean()
```

```python
s2 = df1.groupby('位置')['价格'].sum()
s3 = df1.groupby('位置')['成交量'].sum()
```

```python
#作图
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(121)
ax1.plot(s1.index,s1.values,c= 'c')
ax1.set_xlabel("品牌",fontsize=18)
ax1.set_ylabel('平均价格',fontsize=18,labelpad =12)
plt.xticks(rotation=45)
ax1.xaxis.set_label_coords(1.05,-0.05) # 调整标签位置
ax1.set_title('不同品牌汽车的平均价格')
# 设置位置
ax2 = fig.add_subplot(222)
ax2.plot(s2.index, s2.values,color='c')
ax2.set_xlabel('位置',fontsize =15)
ax2.set_ylabel('成交总量',fontsize =15)
ax2.set_title('不同位置成交总量')
# 设置位置
ax3 = fig.add_subplot(224)
ax3.plot(s3.index, s3.values,color='c')
ax3.set_xlabel('位置',fontsize =15)
ax3.set_ylabel('成交总价',fontsize =15)
ax3.set_title('不同位置成交总价')

# 显示
plt.show()
```
