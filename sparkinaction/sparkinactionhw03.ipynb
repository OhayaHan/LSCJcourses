## sparkinactionhw03
编写一个spark application，并使用cogroup，要求分别实现如下三种场景，将代码上传  
1) 两个RDD都Shuffle  
2) 一个RDD Shuffle另外一个不Shuffle  
3) 两个RDD都不Shuffle  

```python
sc = spark.sparkContext#通过sparkSession获取上下文
```

```python
x = range(100)
```

### 1) 两个RDD都Shuffle
a.Partitioner != cogroupRdd.Partitioner  
b.Partitioner != cogroupRdd.Partitioner  

```python
a=sc.parallelize(x,6).map(lambda x:(x,1))
```

```python
a.getNumPartitions()
```

```python
b = sc.parallelize(range(50,150),6).map(lambda x:(x,1))
```

```python
b.getNumPartitions()
```

```python
rdd4 = a.cogroup(b,4)
```

```python
rdd4.collect()
```

### 2) 一个RDD Shuffle另外一个不Shuffle
a.Partitioner == cogroupRdd.Partitioner  
b.Partitioner != cogroupRdd.Partitioner

```python
a=sc.parallelize(x,4).map(lambda x:(x,1))
b = sc.parallelize(range(50,150),6).map(lambda x:(x,1))
rdd4 = a.cogroup(b,4)
rdd4.collect()
```

### 3) 两个RDD都不Shuffle
a.Partitioner == cogroupRdd.Partitioner  
b.Partitioner == cogroupRdd.Partitioner

```python
a=sc.parallelize(x,4).map(lambda x:(x,1))
b = sc.parallelize(range(50,150),4).map(lambda x:(x,1))
rdd4 = a.cogroup(b,4)
rdd4.collect()
```
