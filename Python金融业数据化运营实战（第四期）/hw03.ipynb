# Q1. 糖果开发商开发制造了一种新型饼干，需要对该厂生产的饼干的水分含量进行置信度为95%的区间估计（数据集为moisture.csv）

利用Python计算95%的置信区间



```python
df = pd.read_csv('moisture.csv')
df.head()
```

```python
mu = df.moisture.mean()
sigma =  df.moisture.std()
#计算区间
stats.t.interval(0.95,len(df)-1,mu,stats.sem(df.moisture))
```

# Q2. 读取EuStockMarkets.csv,包含了欧洲1991到1998年4个主要股票指数收盘价的日度数据，绘制德国DAX指数和英国FTSE指数的散点图，用课程上学的三种相关系数计算这两种股票的相关系数。



```python
df = pd.read_csv('EuStockMarkets.csv')
df.head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Scatter


def scatter_base() -> Scatter:
    c = (
        Scatter()
        .add_xaxis(df['DAX'])
        .add_yaxis("英国FTSE指数", df['FTSE'])
        .set_global_opts(title_opts=opts.TitleOpts(title="Scatter-基本示例"))
    )
    return c
scatter_base().render_notebook()
```

```python
#计算相关系数
print(df[['DAX','FTSE']].corr(method='pearson'))
print(df[['DAX','FTSE']].corr(method='spearman'))
print(df[['DAX','FTSE']].corr(method='kendall'))
```

# Q3. 读取house.csv数据，根据样本数据分析各种因素对购房面积的影响，显著性水平设为0.05

```python
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('house.csv')
df.head()
```

```python
ana = ols('space ~ C(education) + C (unit) + C (income) + C (type)', data = df).fit()
anova_lm(ana)
```

可以看出，income对于购房面积的影响是最为显著的；余下依次为：education， type。unit对于购房面积的影响是不显著的。
