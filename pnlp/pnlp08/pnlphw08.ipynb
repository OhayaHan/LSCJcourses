对train.csv 的数据进行文本聚类，并且计算聚类的正确率  
数据是来自不同新闻主题的新闻报道，尝试对数据进行聚类分析，看能不能把不同主题的新闻区分出来

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df['type'].unique()
```

```python
import jieba
```

```python
df['words'] = df.apply(lambda x:' '.join(jieba.lcut(x[1].strip())),axis=1)
```

```python
df.head()
```

```python
# 定义向量化参数
from sklearn.feature_extraction.text import TfidfVectorizer
#
tfidf_vectorizer = TfidfVectorizer(use_idf=True,  ngram_range=(1,3))
```

```python
docs = list(df['words'])
```

```python
docs[:2]
```

```python
%time tfidf_matrix = tfidf_vectorizer.fit_transform(docs) 
```

```python
tfidf_matrix.shape
```

```python
terms = tfidf_vectorizer.get_feature_names()
terms[:10]
```

```python
##k-means聚类
from sklearn.cluster import KMeans
num_clusters = 10
km = KMeans(n_clusters=num_clusters)
%time km.fit(tfidf_matrix)
```

```python
clusters = km.labels_.tolist()
```

```python
df['cluster'] = clusters
df.head()
```

```python
df1 = df[['type','cluster']]
df1
```

```python
s = df1['cluster']
```

```python
s.value_counts()
```
